```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("tracks.csv")
```

```python
sort_df = df.sort_values("popularity", ascending=True).head(10)
sort_df
```

```python
df.describe().transpose()
```

```python
most_popular=df.query("popularity>90", inplace = False).sort_values("popularity", ascending= False)
most_popular[:10]
```

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index)
df.head(10)
```

```python
df[["name"]].iloc[25]
```

```python
df["duration"]= df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace=True, axis=1)
df.head()
```

```python
corr_df= df.drop(["key", "mode", "explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(15,10))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="icefire", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=30)
```

```python
sample_df = df.sample(int(0.004*len(df)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=sample_df, y="loudness", x="energy", color="b").set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="r").set(title = "popularity vs acousticness Correlation")
```

```python
df["dates"]=df.index.get_level_values("release_date")
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr = df.duration
fig_dims = (15,10)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr=df.duration
sns.set_style(style="whitegrid")
fig_dims = (15,10)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
df_genre =pd.read_csv("SpotifyFeatures.csv")
```

```python
plt.title("Duration of Songs in Different Genres")
sns.color_palette("rocket", as_cmap= True)
fig_dims = (15,10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(y="genre", x="duration_ms", data=df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
plt.show()
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(15,10))
famous = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y="genre", x="popularity", data = famous).set(title="Top 5 Genres by Popularity")
```
